```python
# =========================
# 1. Import Libraries
# =========================
import pandas as pd
import numpy as np
import nltk

# Download required datasets
nltk.download('punkt')
nltk.download('stopwords')

# =========================
# 2. Load Dataset
# =========================
df = pd.read_csv("/content/twitter_validation.csv", encoding="ISO-8859-1", header=None)
df.columns = ["Id", "Location", "Target", "Text"]

# =========================
# 3. Data Cleaning
# =========================

# Remove irrelevant rows
df = df[df["Target"] != "Irrelevant"]

# Reset index
df.reset_index(drop=True, inplace=True)

# Drop unnecessary columns
df.drop(["Id", "Location"], axis=1, inplace=True)

# Convert target labels to numeric
df["Target"] = df["Target"].map({
    'Positive': 1,
    'Negative': -1,
    'Neutral': 0
})

# =========================
# 4. Text Preprocessing
# =========================
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

stop = set(stopwords.words("english"))
stemmer = SnowballStemmer('english')

# Lowercase
tweet = df["Text"].str.lower()

# Remove special characters
tweet = tweet.str.replace(r'[^a-zA-Z\s]', '', regex=True)

# Tokenization + Stopword Removal + Short word removal + Stemming
tweet = tweet.apply(lambda x: [
    stemmer.stem(word)
    for word in word_tokenize(x)
    if word not in stop and len(word) >= 3
])

# Join tokens back to sentence
tweet = tweet.apply(lambda x: " ".join(x))

# =========================
# 5. Vectorization (TF-IDF)
# =========================
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer()
X = vec.fit_transform(tweet)

y = df["Target"].values

# =========================
# 6. Train-Test Split
# =========================
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# =========================
# 7. Model Training
# =========================
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB

models = {
    "Linear SVM": LinearSVC(),
    "Naive Bayes": MultinomialNB()
}

# =========================
# 8. Evaluation
# =========================
from sklearn.metrics import classification_report, accuracy_score

for name, model in models.items():
    print(f"\nModel: {name}")

    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print("====================================")
```
